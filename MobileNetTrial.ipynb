```python
with open("/content/commaSeparated.txt", "r") as myfile:
  text = myfile.read()

myList = text.split(",")
#print(myList)
joined_string = ",".join(myList)
#print(joined_string)
newList = []
for i in myList:
  i = i.replace('"', '')
  if i.startswith('/'):
    continue
  else:
    newList.append(i)

#print(newList)
new_joined_string = ",".join(newList)
#print(new_joined_string)
newString = new_joined_string.replace(",", '","')

with open("finalClasses.txt", "w") as outfile:
    outfile.write(newString)
```

```python
with open("/content/classes.txt", "r") as myfile:
  text = myfile.read()

count = 0
myList = text.split("\n")
print(myList)
newList = []
for i in myList:
  i = i.strip('"')
  if i.startswith('/'):
    count += 1
    continue
  else:
    newList.append(i)
print(newList)
print(count)

with open("DoneClasses.txt", "w") as outfile:
  for i in newList:
    outfile.write(i + "\n")
```
